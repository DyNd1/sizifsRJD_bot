```python
import sqlite3

dbName = "database.db"
tableName = "users"

# Создание соединения с базой данных
conn = sqlite3.connect(dbName,check_same_thread=False)

# Создание курсора для выполнения операций с базой данных
cursor = conn.cursor()

# Функция, добавляющая запись в базу данных
def add_user(id, balance, step, questionIndex, isInQA, isTestMode, name):
    con = sqlite3.connect(dbName)
    cur = con.cursor()
    # SQL-запрос для вставки новой записи
    query = "INSERT INTO users (id, balance, step, questionIndex, isInQA, isTestMode, name) VALUES (?, ?, ?, ?, ?, ?, ?)"

    # Значения, которые будут вставлены в базу данных
    data = (id, balance, step, questionIndex, isInQA, isTestMode, name)

    # Выполнение SQL-запроса с данными
    cur.execute(query, data)

    # Сохранение изменений
    con.commit()
    con.close()

def createTable():
    con = sqlite3.connect(dbName)
    cur = con.cursor()
    cur.execute("CREATE TABLE IF NOT EXISTS {0}(id int(255), balance int(255), step int(255), questionIndex int(255), isInQA int(255), isTestMode int(255), name varchar(255));".format(tableName))
    con.close()

def isInDb(id):
    con = sqlite3.connect(dbName)
    cur = con.cursor()
    querry = """
    SELECT step from {0} where id = "{1}";
    """.format(tableName,id)
    a = cur.execute(querry).fetchone()
    con.close()
    if a == None:
      return False
    return True

def getStepById(id):
    con = sqlite3.connect(dbName)
    cur = con.cursor()
    querry = """
    SELECT step from {0} where id = "{1}";
    """.format(tableName,id)
    a = cur.execute(querry).fetchone()
    con.close()
    return a[0] if a!=None else a

def getBalanceById(id):
    con = sqlite3.connect(dbName)
    cur = con.cursor()
    querry = """
    SELECT balance from {0} where id = "{1}";
    """.format(tableName,id)
    a = cur.execute(querry).fetchone()
    con.close()
    return a[0] if a!=None else a

def getQuestionIndexById(id):
    con = sqlite3.connect(dbName)
    cur = con.cursor()
    querry = """
    SELECT questionIndex from {0} where id = "{1}";
    """.format(tableName,id)
    a = cur.execute(querry).fetchone()
    con.close()
    return a[0] if a!=None else a

def getisInQAById(id):
    con = sqlite3.connect(dbName)
    cur = con.cursor()
    querry = """
    SELECT isInQA from {0} where id = "{1}";
    """.format(tableName,id)
    a = cur.execute(querry).fetchone()
    con.close()
    return a[0] if a!=None else a

def getisTestModeById(id):
    con = sqlite3.connect(dbName)
    cur = con.cursor()
    querry = """
    SELECT isTestMode from {0} where id = "{1}";
    """.format(tableName,id)
    a = cur.execute(querry).fetchone()
    con.close()
    return a[0] if a!=None else a

def setStepById(id,step):
    con = sqlite3.connect(dbName)
    cur = con.cursor()
    querry = """
    UPDATE {0} SET step = {1}  where id = {2};
    """.format(tableName,step,id)
    cur.execute(querry)
    con.commit()
    con.close()

def setQuestionIndexById(id,step):
    con = sqlite3.connect(dbName)
    cur = con.cursor()
    querry = """
    UPDATE {0} SET questionIndex = {1}  where id = {2};
    """.format(tableName,step,id)
    cur.execute(querry)
    con.commit()
    con.close()

def setisTestModeById(id,step):
    con = sqlite3.connect(dbName)
    cur = con.cursor()
    querry = """
    UPDATE {0} SET isTestMode = {1}  where id = {2};
    """.format(tableName,step,id)
    cur.execute(querry)
    con.commit()
    con.close()

def setIsInQAById(id,step):
    con = sqlite3.connect(dbName)
    cur = con.cursor()
    querry = """
    UPDATE {0} SET isInQA = {1}  where id = {2};
    """.format(tableName,step,id)
    cur.execute(querry)
    con.commit()
    con.close()

def setBalanceById(id,step):
    con = sqlite3.connect(dbName)
    cur = con.cursor()
    querry = """
    UPDATE {0} SET balance = {1}  where id = {2};
    """.format(tableName,step,id)
    cur.execute(querry)
    con.commit()
    con.close()



def getQIByName(name):
    con = sqlite3.connect(dbName)
    cur = con.cursor()
    querry = """
    SELECT questionIndex from {0} where name = "{1}";
    """.format(tableName,name)
    a = cur.execute(querry).fetchone()
    con.close()
    return a[0] if a!=None else a



createTable()
add_user(1498479846,100,0,13,0,0,"Лена")
# Закрытие соединения с базой данных
conn.close()
```

```python
import pandas as pd
def checkQestion(user_input) :
    df = pd.read_excel("output.xlsx")
    all_ans = []
    all_quest = []
    for i in range(len(df['answers_merged'])):
        a = eval(df['answers_merged'][i])
        b = eval(df['questions_merged'][i])
        all_ans.append(a)
        all_quest.append(b)
    for i in range(len(all_quest)) :
        if user_input in all_quest[i]:
          return all_ans[i]
        else:
          return GetResponse(user_input)
```
